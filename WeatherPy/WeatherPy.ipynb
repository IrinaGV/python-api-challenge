```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import requests
import time
from scipy.stats import linregress
import seaborn as sns
from scipy import stats

# Import API key
from config import weather_api_key

# Incorporated citipy to determine city based on latitude and longitude
from citipy import citipy

# Output File (CSV)
output_data_file = "output_data/cities.csv"

# Range of latitudes and longitudes
lat_range = (-90, 90)
lng_range = (-180, 180)
```

```python
# List for holding lat_lngs and cities
lat_lngs = []
cities = []

# Create a set of random lat and lng combinations
lats = np.random.uniform(lat_range[0], lat_range[1], size=1500)
lngs = np.random.uniform(lng_range[0], lng_range[1], size=1500)
lat_lngs = zip(lats, lngs)

# Identify nearest city for each lat, lng combination
for lat_lng in lat_lngs:
    city = citipy.nearest_city(lat_lng[0], lat_lng[1]).city_name
    
    # If the city is unique, then add it to a our cities list
    if city not in cities:
        cities.append(city)

# Print the city count to confirm sufficient count
len(cities)
```

```python
url = "http://api.openweathermap.org/data/2.5/weather?"

for city in cities:
    query_url = url + "appid=" + weather_api_key  + "&q=" + city
    weather_dict = requests.get(query_url).json()
    try:
        temp =weather_dict['main']['temp']
                
        print("temperature in " + city + " " + str(weather_dict['main']['temp']))
    
    except Exception as e:
    
        print( "finished ...")   
        
```

```python

print("Beginning Data Retrieval")
print("-----------------------")
for i, city in enumerate(cities, start=1):
    try:

        query_url = url + "appid=" + weather_api_key  + "&q=" + city
        
        weather_dict = requests.get(query_url).json()
        i +=1
        
    except Exception as e:

        print( 'Something broke ...', e)            
    finally:
        print('Processing Record',' ', i,'|', city) 
        
```

```python

name,lat,lon,temp_max,humidity,clouds,wind,country,dt = [],[],[],[],[],[],[],[],[]


for city in cities:
    
    try:
        
        query_url = url + "appid=" + weather_api_key  + "&q=" + city 
        weather_dict = requests.get(query_url).json()
        name.append(weather_dict['name'])
        lat.append(weather_dict['coord']['lat'])
        lon.append(weather_dict['coord']['lon'])
        humidity.append(weather_dict['main']['humidity'])
        wind.append(weather_dict['wind']['speed'])
        temp_max.append(weather_dict['main']['temp_max'])
        clouds.append(weather_dict['clouds']['all'])
        country.append(weather_dict['sys']['country'])
        dt.append(weather_dict['dt'])
       
     
    except Exception as e:
    
        print( "Something broke ...",e)   
    finally:
        print('----')   
        
cities_df = pd.DataFrame({
     'City' :name,
     'Lat':lat,
     'Lng' :lon,
     'Max Temp' :temp_max,
     'Humidity' : humidity,
     'Cloudiness':clouds,
     'Wind Speed' :wind,
     'Country':country,
     'Date':dt,
     }) 

cities_df.head()      
          
    
    
    
    
    
    
    
    
 
```

## Northern Hemisphere - Max Temp vs. Latitude Linear Regression

```python
Filter2=cities_df['Lat']>=0
Cities_df_North = cities_df[Filter2]
Cities_df_North.head(3)
```

```python
x = Cities_df_North['Lat'].astype('float')

y = Cities_df_North['Max Temp'].astype('float')
slope, intercept, r_value, p_value, std_err = stats.linregress(x,y)
print ("r:", r_value)
```

```python

fig = plt.figure(figsize=(8, 6), dpi=80)

ax4 = plt.subplot()

#regression part
slope, intercept, r_value, p_value, std_err = stats.linregress(x,y)

line = slope*x+intercept
plt.plot(x, line, 'r', label='y={:.2f}x+{:.2f}'.format(slope,intercept))
plt.xlabel("Latitude",fontsize=14)
plt.ylabel("Max Temp",fontsize=14)

plt.scatter(x,y, color="b", s=12)
plt.legend(fontsize=6)
plt.legend(loc="lower left", labelcolor  = 'red', prop={'size': 20})


plt.show()
fig.savefig('NhMaxTempVsLat.jpg')
```

The above plot shows a weak negative correlation between Max tempretare and Latitute in Northern Hemisphere on Aug 8th.
